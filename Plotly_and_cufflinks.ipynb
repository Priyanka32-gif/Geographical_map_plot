```python
cf.go_offline()
```

```python
#data
df = pd.DataFrame(np.random.randn(100,4),columns ='A B C D'.split() )
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[33,12,56]})
```

```python
df2
```

```python
df.plot()
```

```python
df.iplot() #interactive plot
```

```python
df.iplot(kind='scatter',x='A',y='B')
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=25)
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],
                    'z':[500,400,300,200,100]})
```
